```python
import numpy as np
import matplotlib.pyplot as plt 

arr = np.arange(1,10001).reshape(100,100)

plt.imshow(arr)
plt.colorbar()
```

```python
%load_ext autotime
mult = np.empty(shape=(10000,10000))
n = 0
for i in range(100):
  for j in range(100):
    v = arr[i][j] * arr
    mult[n,:] = v.ravel()
    n += 1
```

```python
v1 = np.tile(arr.ravel(),(len(arr.ravel()),1)).T
mult_arr_1 = v1*arr.ravel()
```

```python
a = np.ones(10000)
v2 = a*arr.reshape(10000,1)
mult_arr_2 = v2*arr.ravel()
```
